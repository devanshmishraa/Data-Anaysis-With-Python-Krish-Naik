```python
df_grouped = df.groupby('Product_type')['Sales_fillNA'].mean()
df_grouped
```

```python
## Sorting the sales date wise
## First ensure the date column is in datetime format
df['Sales_Date'] = pd.to_datetime(df['Sales_Date'])

df.head()
```

```python
##Now sort the date column in descending order
df_sorted = df.sort_values(by='Sales_Date',ascending=False)
df_sorted
```

```python
## Now let's sort the data by sales column so that we can know the sales value date wise
df_sales_sorted = df.sort_values(by='Sales_fillNA')
df_sales_sorted
```

```python
##Merging and joining the dataframes
#First create the dataframes

df1 = pd.DataFrame({
    'Name':['Dev','Sneh','Sam','Shubh'],
    'Age':[27,26,21,28],
    'Weight':[60,50,55,68]

})

df1
```

```python
df2 = pd.DataFrame({
    'Name':['Dev','Sneh','Ramesh','Suresh'],
    'Age' :[27,26,43,34]
})

df2
```

```python
## Merge Dataframe on the 'Key column'
pd.merge(df1,df2,on='Name',how='inner')
```

```python
pd.merge(df1,df2,on='Name',how='outer')
```

```python
pd.merge(df1,df2,on='Name', how='left')
```
